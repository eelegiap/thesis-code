```python
def initializeRec(recText):
    if isinstance(recText, str):
        return recText.strip()
    else:
        return False

def skipLine(line, idx):
    if len(line.strip()) == 0:
        return False
    
    # throw hashtag line
    if line.strip()[0] == '#':
        return True
    
    # throw attribution line
    for attr in ['из личного','личный блог','источник:','авторский блог']:
        if attr in line.lower():
            return True
        
    # dots at the beginning
    matches = re.search("[\*\+\^-_][*+^-_= ]+", line.strip()) 
    if matches:
        return True
    
    # is none of it alphanumeric
    containsAlpha = False
    for char in line:
        if char.isalpha():
            containsAlpha = True
            break
    if not containsAlpha:
        return True
    
    return False


def processRec(rec):
    recText = initializeRec(rec['Text'])
    cleanLines = []
    if recText:
        # decide which lines to keep
        lines = recText.split('\n')
        for i, line in enumerate(lines):
            if skipLine(line, i):
                continue
            cleanLines.append(line)
    return '\n'.join(cleanLines)
```

```python
%%time
for i, rec in enumerate(records):
    if i % 250 == 0:
        print(f'{i}/{len(records)} parsed')
    cleanedText = processRec(rec)
    rec['cleanedText'] = cleanedText
    rec['doc'] = nlp(cleanedText)
```

```python
%%time
infodict = dict()
infodict['Russian'] = dict()

wordlsts = dict()
poemct = dict()
wordct = dict()
lemmas = dict()
lemmas['Russian'] = set()

for lang in infodict:
    poemct[lang] = dict()
    wordct[lang] = dict()
    for period in ['Before','After']:
        poemct[lang][period]=0
        wordct[lang][period] = 0
        infodict[lang][period] = dict()

random.shuffle(records)
for rec in records:
    lang = 'Russian'
    period = rec['Before or after']
    poemct[lang][period] += 1
    doc = rec['doc']
    lemmaAdded = []
    
    if wordct[lang][period] >= 190000:
        continue
    
    for sent in doc.sentences:
        for token in sent.words:
            wordct[lang][period] += 1

            if token.text.isalpha() and token.text not in rustops and token.text not in ukstops:
                lemma = token.lemma
                lemmas[lang].add(lemma)
                for p in ['Before', 'After']:
                    if lemma not in infodict[lang][p]:
                        infodict[lang][p][lemma] = dict()
                        infodict[lang][p][lemma]['count'] = 0
                        infodict[lang][p][lemma]['examples'] = []
                if lemma in lemmaAdded:
                    continue
                else:
                    infodict[lang][period][lemma]['count'] += 1
                    infodict[lang][period][lemma]['examples'].append({
                        'lineText' : sent.text,
                        'rec' : rec
                    })
                    lemmaAdded.append(lemma)
        
freqdatalist = []
allxvals = set()
allyvals = set()
yvals4SD = []
lang = 'Russian'
for lemma in sorted(list(lemmas[lang])):
    s_1 = infodict[lang]['Before'][lemma]['count']
    s_2 = infodict[lang]['After'][lemma]['count']

    if s_1 != 0 and s_2 != 0:

        sx_ylog = ((math.log(s_1,2)+math.log(s_2,2))/2, math.log(s_2,2) - math.log(s_1,2))
        # sx_yreg = ((s_1+s_2)/2, s_2 - s_1)

        sx_y = sx_ylog
        
        freqdatalist.append({
            'lemma' : lemma,
            's_x' : sx_y[0],
            's_y' : sx_y[1],
            'before' : s_1,
            'after' : s_2,
        })

        allxvals.add(sx_y[0])
        allyvals.add(sx_y[1])
        yvals4SD.append(sx_y[1])
```
